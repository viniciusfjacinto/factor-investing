```python
#%%time
#Abrindo a base
data = pd.read_csv(urlopen(url),encoding='latin-1')
#Retirando sufixo
data['Ativo']=data['Ativo'].str.rstrip('<XBSP>')
#Declarando coluna data como datatime
data['Data'] = pd.to_datetime(data['Data'])
#Renomeando colunas
data.columns = ['ativo','data',"fech","abt","min","max","med","qnegs","qtits","vol"]
```

```python
data1=data1[["ativo", "data", "fech"]]
data1[data1 == "-"] = np.nan
data1=data1[data["data"]>"2001-12-30"]
data1=data1.reset_index()
data1=data1.drop(columns=["index"])
#cot=cot.sort_values(['Ativo', "Data"], ascending=[True, True])
```

```python
data1['ativo'] = data1['ativo'].astype(str)
data1.iloc[:,2:] = data1.iloc[:,2:].astype(float)
data1.dtypes
```

```python
#REARRANJO PARA O MODELO DE FACTORING
balanco4 = data1
balanco4col = list(range(len(balanco4.iloc[:,2:].columns)))
pv = []
for i in balanco4.iloc[:,2:].columns:
    balanco4col = balanco4[['data','ativo',i]]
    balanco4col = balanco4col.pivot_table(values = i, index = 'data', columns = 'ativo', dropna=False)
    pv.append(balanco4col)

listnames = list(balanco4.iloc[:,2:].columns)
dictionary = dict(zip(listnames, pv))
dictionary.keys()
```
